1. **Define Our Mini Knowledge Graph**
Title: CareerAI

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes and relationships for CareerAI
nodes = [
    "User", "Profile", "Skills", "Job Role", "Prompt",
    "Resume", "Cover Letter", "Output", "Feedback", "Job Description"
]
G.add_nodes_from(nodes)

# Add directed egdes to represent relationships
edges = [
    ("User", "Profile"),
    ("Profile", "Skills"),
    ("Skills", "Job Role"),
    ("Job Role", "Prompt"),
    ("Prompt", "Resume"),
    ("Prompt", "Cover Letter"),
    ("Resume", "Output"),
    ("Cover Letter", "Output"),
    ("Output", "Feedback"),
    ("Cover Letter", "Job Description"),
    ("Resume", "Job Description")
]
G.add_edges_from(edges)

plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)

# Draw nodes and edges
nx.draw(G, pos, with_labels=True, node_size=1500, node_color='lightcoral', font_size=10, edge_color="gray", arrows=True)

plt.title("CareerAI Mini Knowledge Graph", fontsize=14)
plt.tight_layout()
plt.show()
```

2. **Build the Graph**

```python
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)

# Draw nodes and edges
nx.draw(G, pos, with_labels=True, node_size=1500, node_color='lightcoral', font_size=10, edge_color="gray", arrows=True)

plt.title("CareerAI Mini Knowledge Graph", fontsize=14)
plt.tight_layout()
plt.show()
```
